#  BREAK THE DATASETS INTO SMALLER DATAFRAME:there is some unwanted columns in the DataFrame so,i have drop it by drop function.



```python
df1.drop(['Confirmed Cases', ' total Cured','total death'],inplace=True,axis=1)
```

```python

df1.head(10)
```

```python
df1.columns=['Month','Region','case',
       'cured', 'death']
```

# CREATING SUBSET OF THE DataFrame BY GROUPBY



```python
g1=df1.groupby('Region')
```

```python
for item,item_df in g1:
    print(item)
```

```python
for item,item_df in g1:
    print(item_df)
```

```python

g1.sum().plot(kind='bar',subplots=True,figsize=(12,7),title='COVID 19 CASE OF DIFFERENT AREA IN INDIA 2020: ')
```

# FROM THE ABOVE GRAPH WE CAN SEE THAT THE MAXIMUM NUMBER OF PEOPLE ARE AFFECTED IN 2020 IS MAHARASHTRA SO WE NEED TO INVESTIGATE IT PROPERLY .


```python

g1.sum().max()
```

```python
g1.sum().max().plot(kind='pie',figsize=(7,9),ylabel='total case')
```

```python
g1.get_group('Maharashtra').plot(kind='scatter',x='case',y='cured',figsize=(14,5),color='r',title='DEPENDENCE BETWEEN NEW CASE AND RECOVER PER DAY')
```

# COMPARISON BETWEEN NEW COVID CASE,CURED AND NUMBER PEOPLE DIE IN 2020 THROUGH OUT THE YEAR IN MAHARASHTRA:


```python
g1.get_group('Maharashtra').plot(kind='line',figsize=(15,5),title='COMPARISION BETWEEN: NEW CASE,CURED AND DEATH')
```

#  FROM THE ABOVE GRAPH OF MAHARASHTRA WE CAN SAY THAT : 1.number of  covid19 cases was increased from  march - september  than start decline.
